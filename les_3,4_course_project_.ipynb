```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```
PATH_TRAIN = '/content/drive/MyDrive/Colab Notebooks/3,4/train.csv'
```

**Data fields**

    Id - идентификационный номер квартиры
    DistrictId - идентификационный номер района
    Rooms - количество комнат
    Square - площадь
    LifeSquare - жилая площадь
    KitchenSquare - площадь кухни
    Floor - этаж
    HouseFloor - количество этажей в доме
    HouseYear - год постройки дома
    Ecology_1, Ecology_2, Ecology_3 - экологические показатели местности
    Social_1, Social_2, Social_3 - социальные показатели местности
    Healthcare_1, Helthcare_2 - показатели местности, связанные с охраной здоровья
    Shops_1, Shops_2 - показатели, связанные с наличием магазинов, торговых центров
    Price - цена квартиры

```
train_df = pd.read_csv(PATH_TRAIN)
train_df.head()
```

1. Построить график распределения целевой переменной. Сделать выводы.

```
plt.figure(figsize = (16, 8))

train_df['Price'].hist(bins=30)
plt.ylabel('Count')
plt.xlabel('Price')

plt.show()
```

2. Построить гистограммы для наиболее проблемных признаков (используйте выводы из первого ДЗ). Сделайте по ним выводы.

```
fig, ax = plt.subplots(2, 2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.hist(train_df['Rooms'])
ax1.set_title('Rooms')

ax2.hist(train_df['KitchenSquare'])
ax2.set_title('KitchenSquare')

ax3.hist(train_df['HouseFloor'])
ax3.set_title('HouseFloor')

ax4.hist(train_df['HouseYear'])
ax4.set_title('HouseYear')

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```
